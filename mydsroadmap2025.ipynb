```python
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches

# Define timeline data
months = ['Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
tasks = [
    ('Excel, SQL, Power BI', 0, 1, 'Core Skills'),
    ('Supervised ML (scikit-learn)', 1, 1, 'ML Foundations'),
    ('Tree-Based Models', 2, 1, 'ML Foundations'),
    ('CNNs, PyTorch, Transfer Learning', 3, 1, 'Deep Learning'),
    ('Prompting, Fine-Tuning, NLP Tasks', 4, 1, 'LLM & NLP'),
    ('Wrap-up, Polish, Apply', 5, 1, 'Portfolio Finalization')
]

# Category colors
colors = {
    'Core Skills': '#4CAF50',
    'ML Foundations': '#2196F3',
    'Deep Learning': '#9C27B0',
    'LLM & NLP': '#FF9800',
    'Portfolio Finalization': '#F44336'
}

# Plot setup
fig, ax = plt.subplots(figsize=(10, 5))
yticks = []
ylabels = []

# Draw tasks as bars
for i, (label, start, duration, category) in enumerate(tasks):
    ax.barh(i, duration, left=start, color=colors[category])
    ax.text(start + duration / 2, i, label, ha='center', va='center', color='white', fontsize=9)
    yticks.append(i)
    ylabels.append(months[start])

# Format chart
ax.set_yticks(yticks)
ax.set_yticklabels(ylabels)
ax.set_xticks(range(len(months)))
ax.set_xticklabels(months)
ax.invert_yaxis()
ax.set_title('DSroadmap2025  Learning Timeline (JulDec 2025)', fontsize=14)
ax.set_xlabel('Month')
ax.set_xlim(0, len(months))

# Legend
patches = [mpatches.Patch(color=col, label=label) for label, col in colors.items()]
ax.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
